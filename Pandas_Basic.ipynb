```python
import pandas as pd
```

```python
pandas_df = pd.read_csv('penguins.csv')
pandas_df
```

<h1>1.apply()</h1>

```python
pandas_df['bill_length_mm'] = pandas_df['bill_length_mm'].apply(lambda x:x/10)
pandas_df
```

<h1>2. nunique()</h1>

```python
pandas_df['species'].nunique()
```

<h1>3. sortvalues()</h1>

```python
pandas_df.sort_values("body_mass_g", ascending=False)
```

```python
pandas_df = pandas_df.rename(columns={'species':'penguin_species'})
pandas_df
```

<h1>5. groupby()</h1>

```python
groupdf = pandas_df.groupby('penguin_species').mean()
```

<h1>6. Query()</h1>

```python
# pandas_df
Adelie_peng = pandas_df.query('penguin_species== "Adelie"')
Adelie_peng
```

```python
melted_df = pandas_df.melt(id_vars = ['penguin_species'], value_vars=["bill_length_mm", "bill_depth_mm"])
melted_df
```

<h1>8.crosstab()</h1>

```python
crosstab = pd.crosstab(pandas_df['penguin_species'],pandas_df['sex'])
crosstab
```
